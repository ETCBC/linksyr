````python
import os
import re
from tf.app import use
````

# Context

````python
P = 'peshitta'
S = 'syrnt'
A = {P: None, S: None}
````

Start the TF browser as follows:

### Old Testament

```
text-fabric peshitta -c --mod=etcbc/linksyr/data/tf/lectio/peshitta
```

### New Testament

```
text-fabric syrnt -c --mod=etcbc/linksyr/data/tf/lectio/syrnt
```

### Example queries

In both cases, issue a query such as

```
verse taksa link
```

or a more refined one:

```
verse taksa link
  word word_etcbc=LLJ>
```


You will see all verses that are associated with a lectionary that has a `taksa` and a `link` value.

````python
for volume in A:
  A[volume] = use(volume, check=True, mod=f'etcbc/linksyr/data/tf/lectio/{volume}')
````

Make TF api names available.

Load extra features

````python
extraFeatures = '''
  lectio
  mark1 mark2
'''.strip().split()

F = {}
T = {}
L = {}

for volume in A:
  thisApi = A[volume].api
  F[volume] = thisApi.F
  T[volume] = thisApi.T
  L[volume] = thisApi.L
  TF = thisApi.TF
  TF.load(extraFeatures, add=True, silent=True)
````

````python
F[P].lectio.freqList()[0:20]
````

````python
def showFreqlist(feature, depth, rep=None):
  theRep = (lambda v: v) if rep is None else rep
  for (value, amount) in feature.freqList()[0:depth]:
    print(f'{amount:>4}x the value {theRep(value)}')
````

````python
showFreqlist(F[P].lectio, 20)
````

````python
def markValue(m):
  parts = m.split('\n')
  return '\n\t' + ('\n\t'.join(parts))
````

````python
showFreqlist(F[P].mark1, 20, markValue)
````

````python
showFreqlist(F[P].mark2, 20, markValue)
````

````python
showFreqlist(F[S].mark1, 20, markValue)
````

````python
showFreqlist(F[S].mark2, 20, markValue)
````

# Queries

First a simple query with all verses with a lectionary (with taksa and link)

````python
query = '''
verse taksa link
'''
````

We run them in both the Old and the New Testament

````python
results = {}
for volume in A:
  results[volume] = A[volume].search(query)
````

Let's show some results from the Old Testament:

````python
A[P].show(results[P], start=1, end=3)
````

Now the New Testament:

````python
A[S].show(results[S], start=1, end=3)
````
